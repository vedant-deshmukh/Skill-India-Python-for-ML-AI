### Loading the Records json file

```python
def load_records():
    file = open("records.json", 'r')
    txt = file.read()
    file.close()

    global records
    records = json.loads(txt)
```

### Defining Global Variables

```python
choice_y =  ['yes', 'ye', 'y']
choice_n = ['no', 'n']
item = 1
quantity = 1
buy_id = []
buy_qn = []
buy_pr = []
t_id = 0
bill_time = 1
sales = {}
records ={}
```

### Reset Function

This function will reset the cart when a transaction for the bill is completed.</h4>

```python
def reset():
    file = open("records.json", 'r')
    txt = file.read()
    file.close()

    global records
    records = json.loads(txt)
    global buy_id
    buy_id = []
    global buy_qn
    buy_qn = []
    global buy_pr  
    buy_pr = []
    global bill_time
    bill_time = 1
    global sales 
    sales ={}
```

### Item Check Function

This function is used to check if a product id is present in the records or not.

```python
def item_check():
    global item
    item = str(input("Product ID : "))
    if item in records:
        print("ID Matched!")
    else:
        print("Invalid ID. Try Again!")
        item_check()
```

### Quantity Check Function

This function is used at the time of buying items, to check if a product is in the stock or not.

```python
def quantity_check():
    global quantity
    quantity = int(input("Enter Quantity : "))
    if (quantity <= records[item]['stock']):
        buy_id.append(item)
        buy_qn.append(quantity)
    else:
        print("Less No. of Items Available!. Try Again!")
        quantity_check()
```

### Function to View Items in Inventory

This function displays the list of items with their price.

```python
def view_items():
    print("----------------------------------------ITEMS LIST--------------------------------------------------")
    print("\n-----Following are the names and prices (per unit) of items in Inventory-----")
    for key in records:
        print("------------------")
        print("Product ID :      " + str(key))
        print("Name :      " + str(records[key]['name']))
        print("Price :     " + str(records[key]['price']))
    print("\n")
```

### Add Items Function

This function is used to add new items to the inventory or upgrade the stock of existing items in the inventory.

```python
def add_items():
    z = 0
    print("********************ADD NEW ITEMS TO INVENTORY***********************")
    print("\n")
    choice = 'yes'
    flag =  True
    while(flag == True):
        choice  = input("Do you want to add a Item? ").lower()
        if choice in choice_y:
            prod_id = str(input("Enter product id:"))
            if prod_id in records.keys():
                qn = int(input("Enter Quantity:"))    
                records[prod_id]['stock'] =  qn
                z = z + 1
            else:
                name = str(input("Enter Name : "))
                typ = str(input("Enter Type : "))
                pr = int(input("Enter Price:"))
                qn = int(input("Enter Quantity:"))
                exp = str(input("Enter Expiry Date (DD/MM/YYYY) : "))
                records[prod_id] = {'name': name, 'type': typ, 'price': pr, 'stock': qn, 'exp_date' : exp}
                z = z + 1
        elif choice in choice_n:
            print( str(z) +" Items Added to Inventory!\n Thank You!")
            print("\n")
            flag = False
            break

    new = json.dumps(records)
    files = open("records.json",'w')
    files.write(new)
    files.close()
    
    print("***************************************************************************")
```

### Delete Items Function

This function is used to delete items from inventory.

```python
def delete_items():
    while(True):
        ch =  input("Do you want to delete a item? ").lower()
        if ch in choice_y:
            i = input("Enter the Product ID : ")
            if i in records:
                del records[i]
                print("Item Deleted Successfully!")
                new = json.dumps(records)
                files = open("temp.json",'w')
                files.write(new)
                files.close()
            else: 
                print("ID not Found!")
                print("\n")
                delete_items()
        if ch in choice_n:
            print("Thank You!")
            break
    
    new = json.dumps(records)
    files = open("records.json",'w')
    files.write(new)
    files.close()
```

### Buy Items Function

This function is used when a customer wants to buy some items.

```python
def buy_items():
    print("**--**--**--**--**--**--WELCOME TO YOUR CART**--**--**--**--**--**--**--**")
    print("\n")
    flag  = True
    count = 1
    while(flag == True): 
        choice  = input("Do you want to buy a Item? ").lower()
        if choice in choice_y: 
            print("\n--------#ITEM "+ str(count)+ " ---------" )
            item_check()
            quantity_check()
            count = count + 1
            print("Item Added Successfully!")
            print("\n")
        elif choice in choice_n:
            print("***************************************************************************")
            print("******GENERATING YOUR BILL......******")
            print("\n")
            flag = False
            generate_bill()
        else:
            print("Invalid Input. Try Again!")
            pass
        
```

### Generate Bill Function

This function helps to generate bill for the customer.

```python
def generate_bill():
    print("**--**--**--**--**--**--PURCHASE BILL**--**--**--**--**--**--**--**")
    i = 0
    global bill_time
    bill_time = datetime.now()
    print("\nBill Generated On : " + str(bill_time) )
    print("\n")
    while (i<len(buy_id)):
        print("ITEM " + str(i+1))
        print("----------------")
        print("Product ID :      " +  str(buy_id[i]))
        print("Name :            " + (records[buy_id[i]]['name'] ))
        print("Price(per unit) : " + str(records[buy_id[i]]['price']) )
        print("Quantity :        " + str(buy_qn[i]) )
        cost =  buy_qn[i] * int(records[buy_id[i]]['price'])
        buy_pr.append(cost)
        print ("Amount :          " + str(cost))
        print("\n")
        i =  i + 1
    total = sum(buy_pr)
    print("Your Total Billed Amount is : " +str(total))
    update_after_bill()
    save_sales()
    reset()
```

### Function to update Inventory

This function updates the inventory when a transaction is completed by customer.

```python
def update_after_bill():
    i = 0
    while(i< len(buy_id)):
        diff = int(records[buy_id[i]]['stock']) - buy_qn[i] 
        records[buy_id[i]]['stock'] = diff
        i = i + 1
        
        
    new = json.dumps(records)
    files = open("records.json",'w')
    files.write(new)
    files.close()
    
    
```

### Function to maintain Sales records

This function is responsible for maintaing the sales records

```python
def save_sales():
    i = 0
    str_time = str(bill_time)
    while (i < len(buy_id)):
        global t_id
        t_id = t_id + 1
        temp =  int(records[buy_id[i]]['price']) * int(buy_qn[i])
        global sales
        sales[t_id] =   {'product_id': buy_id[i],
                         'name': records[buy_id[i]]['name'],
                         'price' : records[buy_id[i]]['price'],
                         'quantity': buy_qn[i],
                         'amount' : temp,
                         'time': str_time   }
                
        i = i + 1
     
    new = json.dumps(sales)
    direct =  os.path.dirname(os.path.realpath("sales.json"))
    files = os.listdir(direct)
    if "sales.json" in files:
        files = open("sales.json",'a')
        files.write(new)
        files.truncate()
        files.close()
    else:
        files = open("sales.json",'w')
        files.write(new)
        files.close() 
```

### Functions that help user to interact with the Software

These functions are responsible for providing user the interface to interact with the software.

```python
def switch():
    char = input("Enter Your Choice : ").lower()
    print("\n")
    
    def view():
        view_items()
    def add():
        add_items()
    def delete():
        delete_items()
    def buy():
        buy_items()
    def bye():
        print("Thank You! See you Soon!")
    def default():
        print("Invalid Input!")

    dict = {
        'a' : view,
        'b' : add,
        'c' : delete,
        'd': buy,
        'e': bye   }
    
    dict.get(char,default)()
```

```python
def start():
    load_records()
    print("************************WELCOME TO THIS INVENTORY MANAGEMENT SYSTEM************************")
    print("\nPress A for viewing Items currently in Inventory")
    print("Press B for Adding Items to Inventory")
    print("Press C for Deleting Items from Inventory")
    print("Press D for Buying Some Items")
    print("Press E to Exit")
    
    
    switch()
```

```python
def again():
    while(True):
        char = input("\nDo you wish to continue? ").lower()
        if char in choice_y:
            print("\n")
            start()
        if char in choice_n:
            print("BYE !!!!!")
            break
        else:
            print("Invalid Input. Try Again!")
            again()
```

### Function to start the IMS

This function will start the software.

```python
def start_software():
    start()
    again()
```

<b>For User</b> : Type start_software() to start working on the software.
